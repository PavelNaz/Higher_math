```python
x = []
y1 = []
y2 = []
r = 10
for i in range (-1000, 1000):
    x_ = i/50
    x.append(x_)
    y1.append(np.sqrt(r**2 - x_**2))
    y2.append(-np.sqrt(r**2 - x_**2))
fig, ax = plt.subplots(1)
ax.set_aspect(1)
plt.plot(x, y1, color='red')
plt.plot(x, y2, color='red')
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = []
y_e1 = []
y_e2 = []
el1 = 2
el2 = 1
for i in range (-1000, 1000):
    x_ = i/100
    x.append(x_)
    y_e1.append(np.sqrt(el1**2-(x_*el1/el2)**2))
    y_e2.append(-np.sqrt(el1**2-(x_*el1/el2)**2))
fig, ax = plt.subplots(1)
ax.set_aspect(1)
plt.plot(x, y_e1, color='red')
plt.plot(x, y_e2, color='red')
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = []
y_g1 = []
y_g2 = []
g_1 = 12
g_2 = 2
for i in range (-1000, 1000):
    x_ = i/120
    x.append(x_)
    y_g1.append(np.sqrt(g_1**2+(g_1*x_/g_2)**2))
    y_g2.append(-np.sqrt(g_1**2+(g_1*x_/g_2)**2))  
plt.plot(x, y_g1, color='red')
plt.plot(x, y_g2, color='red')
plt.xlabel("x")
plt.ylabel("y")
```

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 100)
y = x**2 - 1
plt.plot(x, y, color='red')

x1 = np.linspace(-5, 5, 100)
y1 = (- 1 + np.exp(x1) + x1)/x1
plt.plot(x1, y1, color='blue')

plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
plt.grid(True)
plt.show()
```

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1., np.exp(x) + x * (1. - y) - 1.)

x1, y1 = (fsolve(equations, (-2, 10)))
x2, y2 =(fsolve(equations, (0, 0)))
x3, y3 =(fsolve(equations, (2, 10)))
x4, y4 =(fsolve(equations, (4, 10)))

print("%.2f"%x1, "%.2f"%y1)
print("%.2f"%x2, "%.2f"%y2)
print("%.2f"%x3, "%.2f"%y3)
print("%.2f"%x4, "%.2f"%y4)
```

```python
def f(p):
    x = p
    return (np.exp(x)-x**3+2*x-1)
print("Уравнения системы принимают нулевые значения при значениях х:")
x1 = fsolve(f, -2)
print("%.2f"%x1)
x2 = fsolve(f, 0)
print("%.2f"%x2)
x3 = fsolve(f, 2)
print("%.2f"%x3)
x4 = fsolve(f, 4)
print("%.2f"%x4)
print("Значения функции по условиям системы должны быть больше нуля.")
print(f" Функция принимает положительные значения в диапазонах: (-{np.inf};{x1}), ({x2}, {x3}), ({x4}, {np.inf})")
x = np.linspace(-3, 10, 500)
plt.plot([-5, 10], [0, 0], c="red", linewidth=0.8, linestyle='--')
y = np.exp(x)-x**3+2*x-1
plt.plot(x, y, label='y=(exp(x)-x^3+2*x-1', c='red')
plt.ylim(-5, 10)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
